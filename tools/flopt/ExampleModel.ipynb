## Definitions

```python
from gekko import GEKKO

g = GEKKO(remote=False)

max_ratio = 4

def SMAX(x, y, sroot: GEKKO, tau = 0.01):
    return 0.5 * (x + y + sroot( (x - y)**2 + 4 * tau * tau ))
    
def THIN(ratio):
    return ratio + 1/ratio - 1

x1, y1, w1, h1 = g.Var(), g.Var(), g.Var(), g.Var()
x2, y2, w2, h2 = g.Var(), g.Var(), g.Var(), g.Var()
```

## Initialization

```python
x1.value, y1.value, w1.value, h1.value = 2, 4, 3, 5
x2.value, y2.value, w2.value, h2.value = 2, 2, 4, 4

A1 = 5
A2 = 16
```

## Model Definition

```python
# Model
g.Equation(w1 > 0)
g.Equation(h1 > 0)
g.Equation(w2 > 0)
g.Equation(h2 > 0)

g.Equation(THIN(w1 / h1) <= THIN(max_ratio))
g.Equation(THIN(w2 / h2) <= THIN(max_ratio))

g.Equation(w1 * h1 >= A1)
g.Equation(w2 * h2 >= A2)

t1 = (x1 - x2)**2 - 0.25 * (w1 + w2)**2
t2 = (y1 - y2)**2 - 0.25 * (h1 + h2)**2
g.Equation(SMAX(t1, t2, g.sqrt) >= 0)
```

## Solve

```python
g.solve(disp=True)
```
